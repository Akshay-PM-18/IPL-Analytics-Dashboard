```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully!")
```

```python
matches = pd.read_csv("../data/raw/matches.csv")
deliveries = pd.read_csv("../data/raw/deliveries.csv")

print("Datasets loaded successfully!")
```

```python
print("Matches shape:", matches.shape)
print("Deliveries shape:", deliveries.shape)
```

```python
print("MATCHES COLUMNS:\n")
print(matches.columns.tolist())

print("\nDELIVERIES COLUMNS:\n")
print(deliveries.columns.tolist())
```

```python
print("MATCHES HEAD:\n")
display(matches.head())

print("\nDELIVERIES HEAD:\n")
display(deliveries.head())
```

```python
matches.info()
deliveries.info()
```

```python
matches_clean = matches.copy()
deliveries_clean = deliveries.copy()

print("Working copies created successfully!")
```

```python
print("Missing values in MATCHES:\n")
print(matches_clean.isnull().sum())

print("\nMissing values in DELIVERIES:\n")
print(deliveries_clean.isnull().sum())
```

```python
# convert season to numeric
matches_clean["season"] = pd.to_numeric(matches_clean["season"], errors="coerce")

# convert date to datetime
matches_clean["date"] = pd.to_datetime(matches_clean["date"])

print("Data types cleaned successfully!")
```

```python
deliveries_clean["over"] = deliveries_clean["over"] + 1

print("Over column corrected successfully!")
```

```python
deliveries_clean["over"].min(), deliveries_clean["over"].max()
```

```python
print("Unique teams in MATCHES (team1):")
print(sorted(matches_clean["team1"].unique()))

print("\nUnique teams in DELIVERIES (batting_team):")
print(sorted(deliveries_clean["batting_team"].unique()))
```

```python
team_name_mapping = {
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru",
    "Rising Pune Supergiants": "Rising Pune Supergiant"
}

print("Team name mapping created successfully!")
```

```python
# matches table columns
matches_clean["team1"] = matches_clean["team1"].replace(team_name_mapping)
matches_clean["team2"] = matches_clean["team2"].replace(team_name_mapping)
matches_clean["toss_winner"] = matches_clean["toss_winner"].replace(team_name_mapping)
matches_clean["winner"] = matches_clean["winner"].replace(team_name_mapping)

# deliveries table columns
deliveries_clean["batting_team"] = deliveries_clean["batting_team"].replace(team_name_mapping)
deliveries_clean["bowling_team"] = deliveries_clean["bowling_team"].replace(team_name_mapping)

print("Team names standardized successfully!")
```

```python
print("Unique teams in MATCHES after cleaning:")
print(sorted(matches_clean["team1"].unique()))

print("\nUnique teams in DELIVERIES after cleaning:")
print(sorted(deliveries_clean["batting_team"].unique()))
```

```python
# save cleaned matches dataset
matches_clean.to_csv("../data/processed/matches_clean.csv", index=False)

# save cleaned deliveries dataset
deliveries_clean.to_csv("../data/processed/deliveries_clean.csv", index=False)

print("Cleaned datasets saved successfully!")
```

```python
ipl_df = deliveries_clean.merge(
    matches_clean,
    left_on="match_id",
    right_on="id",
    how="left"
)

print("Merged dataset created successfully!")
print("Merged dataset shape:", ipl_df.shape)
```

```python
def get_phase(over):
    if over <= 6:
        return "Powerplay"
    elif over <= 15:
        return "Middle Overs"
    else:
        return "Death Overs"

ipl_df["phase"] = ipl_df["over"].apply(get_phase)

print("Phase column created successfully!")
```

```python
print(ipl_df["phase"].value_counts())
```

```python
ipl_df["is_boundary"] = ipl_df["batsman_runs"].apply(lambda x: 1 if x in [4, 6] else 0)

print("Boundary column created successfully!")
```

```python
print(ipl_df["is_boundary"].value_counts())
```

```python
ipl_df[["batsman_runs", "is_boundary"]].head(10)
```

```python
ipl_df["is_dot_ball"] = ipl_df["total_runs"].apply(lambda x: 1 if x == 0 else 0)

print("Dot ball column created successfully!")
```

```python
print(ipl_df["is_dot_ball"].value_counts())
```

```python
ipl_df[["total_runs", "is_dot_ball"]].head(10)
```

```python
top_batters = (
    ipl_df.groupby("batter")["batsman_runs"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_batters)
```

```python
ipl_df["is_legal_ball_faced"] = ipl_df["extras_type"].apply(lambda x: 0 if x == "wides" else 1)

print("Legal ball faced column created successfully!")
```

```python
batter_stats = (
    ipl_df.groupby("batter")
    .agg(
        total_runs=("batsman_runs", "sum"),
        balls_faced=("is_legal_ball_faced", "sum"),
        boundaries=("is_boundary", "sum")
    )
    .reset_index()
)

batter_stats["strike_rate"] = (batter_stats["total_runs"] / batter_stats["balls_faced"]) * 100
batter_stats["boundary_percent"] = (batter_stats["boundaries"] / batter_stats["balls_faced"]) * 100

batter_stats.head()
```

```python
batter_stats_filtered = batter_stats[batter_stats["balls_faced"] >= 300].copy()

print("Filtered batter stats shape:", batter_stats_filtered.shape)
batter_stats_filtered.head()
```

```python
top_strike_rate_batters = (
    batter_stats_filtered.sort_values("strike_rate", ascending=False)
    .head(10)
)

top_strike_rate_batters[["batter", "total_runs", "balls_faced", "strike_rate", "boundary_percent"]]
```

```python
ipl_df["is_legal_ball_bowled"] = ipl_df["extras_type"].apply(
    lambda x: 0 if x in ["wides", "noballs"] else 1
)

print("Legal ball bowled column created successfully!")
```

```python
bowler_wicket_types = [
    "bowled",
    "caught",
    "lbw",
    "stumped",
    "caught and bowled",
    "hit wicket"
]

ipl_df["is_bowler_wicket"] = ipl_df["dismissal_kind"].apply(
    lambda x: 1 if x in bowler_wicket_types else 0
)

print("Bowler wicket column created successfully!")
```

```python
bowler_stats = (
    ipl_df.groupby("bowler")
    .agg(
        wickets=("is_bowler_wicket", "sum"),
        runs_conceded=("total_runs", "sum"),
        balls_bowled=("is_legal_ball_bowled", "sum"),
        dot_balls=("is_dot_ball", "sum"),
        boundaries_conceded=("is_boundary", "sum")
    )
    .reset_index()
)

bowler_stats["overs_bowled"] = bowler_stats["balls_bowled"] / 6
bowler_stats["economy"] = bowler_stats["runs_conceded"] / bowler_stats["overs_bowled"]
bowler_stats["dot_ball_percent"] = (bowler_stats["dot_balls"] / bowler_stats["balls_bowled"]) * 100

bowler_stats.head()
```

```python
bowler_stats_filtered = bowler_stats[bowler_stats["balls_bowled"] >= 300].copy()

print("Filtered bowler stats shape:", bowler_stats_filtered.shape)
bowler_stats_filtered.head()
```

```python
top_wicket_takers = (
    bowler_stats_filtered.sort_values("wickets", ascending=False)
    .head(10)
)

top_wicket_takers[["bowler", "wickets", "economy", "dot_ball_percent"]]
```

```python
best_economy_bowlers = (
    bowler_stats_filtered.sort_values("economy", ascending=True)
    .head(10)
)

best_economy_bowlers[["bowler", "wickets", "economy", "dot_ball_percent"]]
```

```python
team_batting_stats = (
    ipl_df.groupby("batting_team")
    .agg(
        total_runs=("total_runs", "sum"),
        total_balls=("is_legal_ball_faced", "sum"),
        wickets_lost=("is_wicket", "sum"),
        boundaries=("is_boundary", "sum"),
        dot_balls=("is_dot_ball", "sum")
    )
    .reset_index()
)

team_batting_stats["strike_rate"] = (team_batting_stats["total_runs"] / team_batting_stats["total_balls"]) * 100
team_batting_stats["boundary_percent"] = (team_batting_stats["boundaries"] / team_batting_stats["total_balls"]) * 100
team_batting_stats["dot_ball_percent"] = (team_batting_stats["dot_balls"] / team_batting_stats["total_balls"]) * 100

team_batting_stats
```

```python
top_scoring_teams = (
    team_batting_stats.sort_values("total_runs", ascending=False)
    .head(10)
)

top_scoring_teams[["batting_team", "total_runs", "strike_rate", "boundary_percent", "dot_ball_percent"]]
```

```python
# total matches played per team
matches_played_team1 = matches_clean["team1"].value_counts()
matches_played_team2 = matches_clean["team2"].value_counts()

total_matches_played = matches_played_team1.add(matches_played_team2, fill_value=0)

# total matches won
matches_won = matches_clean["winner"].value_counts()

# create dataframe
team_results = pd.DataFrame({
    "matches_played": total_matches_played,
    "matches_won": matches_won
}).fillna(0)

team_results["win_percentage"] = (team_results["matches_won"] / team_results["matches_played"]) * 100

team_results = team_results.reset_index().rename(columns={"index": "team"})

team_results
```

```python
top_teams = team_results.sort_values("win_percentage", ascending=False).head(10)

top_teams
```

```python
team_phase_stats = (
    ipl_df.groupby(["batting_team", "phase"])
    .agg(
        total_runs=("total_runs", "sum"),
        total_balls=("is_legal_ball_faced", "sum"),
        boundaries=("is_boundary", "sum"),
        dot_balls=("is_dot_ball", "sum")
    )
    .reset_index()
)

team_phase_stats["strike_rate"] = (team_phase_stats["total_runs"] / team_phase_stats["total_balls"]) * 100
team_phase_stats["boundary_percent"] = (team_phase_stats["boundaries"] / team_phase_stats["total_balls"]) * 100
team_phase_stats["dot_ball_percent"] = (team_phase_stats["dot_balls"] / team_phase_stats["total_balls"]) * 100

team_phase_stats.head(15)
```

```python
# filter only current teams and only Powerplay
powerplay_chart = team_phase_stats[
    (team_phase_stats["batting_team"].isin(current_teams)) &
    (team_phase_stats["phase"] == "Powerplay")
].sort_values("total_runs", ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=powerplay_chart, x="total_runs", y="batting_team")

plt.title("IPL Current Teams - Powerplay Runs")
plt.xlabel("Total Runs in Powerplay")
plt.ylabel("Team")

# save chart
plt.savefig("../outputs/current_teams_powerplay_runs.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# filter current teams + Death Overs
death_overs_chart = team_phase_stats[
    (team_phase_stats["batting_team"].isin(current_teams)) &
    (team_phase_stats["phase"] == "Death Overs")
].sort_values("strike_rate", ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=death_overs_chart, x="strike_rate", y="batting_team")

plt.title("IPL Current Teams - Death Overs Strike Rate")
plt.xlabel("Strike Rate (Death Overs)")
plt.ylabel("Team")

# save chart
plt.savefig("../outputs/current_teams_death_overs_strike_rate.png", dpi=300, bbox_inches="tight")

plt.show()
```
